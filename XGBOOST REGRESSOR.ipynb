```python
dataset = pd.read_csv('insurance_pre.csv')
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)  #data pre process(converting string into numner-nominal data)
```

```python
dataset
```

```python
dataset.columns
```

```python
indep=dataset[['age', 'bmi', 'children','charges','sex_male', 'smoker_yes']]
```

```python
indep
```

```python
depend=dataset[['charges']]
```

```python
depend
```

```python
from sklearn.model_selection import train_test_split             #sklearn library used to split train set and test set 

X_train,X_test,Y_train,Y_test=train_test_split(indep, depend, test_size=0.30, random_state=0)

                                                              #Then call train_test_split function it is parameterised function
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train
```

```python
import xgboost as xg
regressor = xg.XGBRegressor(n_estimators=300, learning_rate=0.05, max_depth=4, subsample=0.9, colsample_bytree=0.7, reg_alpha=0.1, random_state=42)

regressor.fit(X_train,Y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,y_pred)                                         #To find the r2_score,we take actual output and predictor.
```

```python
r_score
```

```python
import pickle                             #Pickle is uesd for save the modle(pickle filename extention is 'sav')
filename="finalized_model_XGboost.sav"        # creat file    
```

```python
pickle.dump(regressor, open(filename,"wb"))  #modle saved in that file
```

```python
loaded_modle=pickle.load(open("finalized_model_XGboost.sav", "rb")) # load the modle (Rb means reading)
result=loaded_modle.predict([[300,400,300,1,0]])
```
